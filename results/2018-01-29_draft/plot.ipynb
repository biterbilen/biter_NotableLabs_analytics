```python
%matplotlib inline
import matplotlib
#matplotlib.use('Agg')
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
dta = sns.load_dataset("tips")
dta_sample = dta.sample(100)
dta_sample
sns.set(style="white")
g = sns.PairGrid(dta)
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter, alpha=0.4, marker=".")
g.add_legend()
#g.savefig("foo.pdf")

```

```python
g = sns.FacetGrid(dta, col="time")
g.map(plt.hist, "tip")
```

```python
dta.columns.names
g = sns.FacetGrid(dta, col="sex", hue="smoker", margin_titles=True)
g.map(plt.scatter, "total_bill", "tip", alpha=.7)
g.add_legend()
```
