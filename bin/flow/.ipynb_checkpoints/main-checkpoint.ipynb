```python
%run visualize_data
title = 'within-well effect for raw data:'
x = "wellid"
for y in dta.columns.values:
    if y in ['event', 'name', 'label', 'wellid']:
        continue
    odir  = 'raw_violin_' + x + '/'
    ofile = odir + 'raw_violin_' + x + '_vs_' + y + '.pdf'
    #violin(dta, ofile, x=x, y=y, title=title)
    #break
# summarize data per/well to check the data range (confirm visualizations)
dta.drop(['event', 'label', 'name'], axis=1).groupby(by=['wellid']).describe()

# CONC: data is mostly right tailed and also with negative values - data transformation necessary
```

```python
%run transform_data
channels = dta.columns.values[[x not in ['event', 'name', 'label', 'wellid', 'Time'] for x in dta.columns.values]]
#print channels
tdta = dta.copy()
tdta.loc[:,channels] = transform(tdta, type='hlog', channels=channels)
print dta.shape, tdta.shape

%run visualize_data
title = 'within-well effect for hlog-transformed data:'
x = "wellid"
for y in tdta.columns.values:
    if y in ['event', 'name', 'label', 'wellid']:
        continue
    odir  = 'hlog_violin_' + x + '/'
    ofile = odir + 'hlog_violin_' + x + '_vs_' + y + '.pdf'
    #violin(tdta, ofile, x=x, y=y, title=title)
# summarize data per/well to check the data range (confirm visualizations)
tdta.drop(['event', 'label', 'name'], axis=1).groupby(by=['wellid']).min()

# CONC: Data is still packed except for FSC-H and FSC-A
```

```python
%run transform_data
channels = dta.columns.values[[x not in ['event', 'name', 'label', 'wellid', 'Time'] for x in dta.columns.values]]

tdta2 = dta.copy()
tdta2.loc[:,channels] = transform(tdta2, type='arcsinh', channels=channels)
print dta.shape, tdta2.shape

%run visualize_data
title = 'within-well effect for arcsinh-transformed data:'
x = "wellid"
for y in tdta.columns.values:
    if y in ['event', 'name', 'label', 'wellid']:
        continue
    odir  = 'arcsinh_violin_' + x + '/'
    ofile = odir + 'arcsinh_violin_' + x + '_vs_' + y + '.pdf'
    #violin(tdta2, ofile, x=x, y=y, title=title)
# summarize data per/well to check the data range (confirm visualizations)
tdta2.drop(['event', 'label', 'name'], axis=1).groupby(by=['wellid']).describe()

# CONC: Data is better distributed with arcsinh
# CONC: No need for time dependent normalization
```

```python
# Check which factors are informative for cell label
%run visualize_data
title = 'label effect for arcsinh-transformed data:'
x = "label"
for y in tdta.columns.values:
    if y in ['event', 'name', 'label', 'wellid']:
        continue
    odir  = 'arcsinh_violin_' + x + '/'
    ofile = odir + 'arcsinh_violin_' + x + '_vs_' + y + '.pdf'
    #violin(tdta2, ofile, x=x, y=y, title=title)
# summarize data per/label to check the data range (confirm visualizations)
tdta2.drop(['event', 'wellid', 'name'], axis=1).groupby(by=['label']).describe()

# CONC: Half of the data is debris, ~33% live and ~17% dead cells
# CONC: Potential discriminatory factors: CD19, CD33, CD34, DAPI, FSC, SSC, VL3, VL5
```
